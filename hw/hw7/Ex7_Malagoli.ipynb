# Ising Model #

Consider $( N $) spin-1/2 particles on a one-dimensional lattice, described by the Hamiltonian:  

$$
\hat{H} = \lambda \sum_{i}^{N} \sigma_z^i + \sum_{i}^{N-1} \sigma_x^i \sigma_x^{i+1}
$$

where $( \sigma_x )$ and $( \sigma_z $) are the Pauli matrices, and $( \lambda $) is the interaction strength.

#### Tasks:  
1. **Hamiltonian Matrix Representation:**  
   - Write a program to compute the $( N \times N $) matrix representation of the Hamiltonian $( \hat{H} $) for different values of $( N $).  

2. **Diagonalization:**  
   - Diagonalize $( H $) for different $( N = 1, \dots, N_{\text{max}} $) and $( \lambda \in [0, -3] $).  
   - Determine the largest $( N_{\text{max}} $) you can reach.  

3. **Spectrum Analysis:**  
   - Plot the first $( k $) energy levels as a function of $( \lambda $) for different $( N $).  
   - Provide comments on the spectrum.  

```python
import ising as ig
import numpy as np
```

## Exercise 2 - Find N_max ##

```python
# Parameters
N_values = np.arange(4, 14, 1)  # Number of spins
lam = 1  

# Normal implementation
for N in N_values:
    print(f"Calculating for N = {N}")
    ham = ig.construct_hamiltonian(N, lam)
    ig.diagonalize_hamiltonian(ham, verb=2)   
```

```python
# Parameters
N_values = np.arange(4, 20, 1)  # Number of spins
lam = 1  

# Sparse matrices implementation
for N in N_values:
    print(f"Calculating for N = {N}")
    ham = ig.sparse_ising(N, lam)
    ig.diagonalize_hamiltonian(ham, mode = 'sparse', verb=2)  
```

As can be seen from above, given the available computation resources, the maximum value of N achievable (considering an acceptable computation time to be under 5 minutes) through the 'normal' implementation is $N=13$. On the other hand, through the sparse matrices implementation, the maximum value of N achievable is $N=15$. Therefore, it can be seen how, by using sparse matrices, we can more comfortably reach higher values for N. Obviously, the other advantage is memory usage: for dense matrices, at $N = 14$ we need around 1 Gb, while for sparse matrices, we only need 3.45 Mb.
